```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Question 1) Please read the "happiness.csv" file as a DataFrame object named as "happiness_df". Print the length and the first 5 tuples of the DataFrame. (2 points)

```python
#codes
happiness = pd.read_csv('happiness.csv')
happiness_df = pd.DataFrame(happiness)
print("Length of the DataFrame:",happiness.shape)
happiness_df.head()
```

Question 2) Print the top 20 countries by generosity value in descending order. (5 points)

```python
#codes
sorted_happiness = happiness_df.sort_values( by =['Generosity'], ascending = [False])
sorted_happiness.head(20)
```

Question 3) Check how many rows null in the data (3 points)

```python
#codes
count = 0
for i in range(len(happiness_df.index)) :
    if happiness_df.iloc[i].isnull().sum()>0:
        count= count+1
    
count
```

Question 4) Delete rows which have any null value (5 points)

```python
#codes
happiness_df = happiness_df.dropna(how='any',axis=0) 
happiness_df.head(21)
```

Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
#codes
conditions = [(happiness_df['Ladder']< 30),(30 <= happiness_df['Ladder']) &(happiness_df['Ladder']< 60),(60 <= happiness_df['Ladder']) & (happiness_df['Ladder'] < 90),(90 <= happiness_df['Ladder']) & (happiness_df['Ladder'] < 120),(happiness_df['Ladder'] >= 120)]
choices = ['very happy','happy','moderately happy','unhappy','very unhappy']

happiness_df['happiness_degree'] = np.select(conditions, choices, default=0)
happiness_df.head(10)
```

Question 6) Create heat map and comment which values are correlated with each other (15 points)

```python
#codes
corr_table = happiness_df.corr()
plt.figure(figsize=(20, 15))
heat_map = sns.heatmap(corr_table,annot=True,yticklabels=True)
plt.yticks(rotation=0)
sns.set(font_scale=1.6)
# Positive affect and SD of ladder attributes has no correlation.
# Social support,log of GDP per capita and healthy life expectancy attributes have strongly positive correlation with ladder.
# Social support is correlated with log of GDP per capita.
# Log of GDP per capita has correlation with healthy life expectancy.
```

Question 8) Display mean values of columns according to happiness_degree categories (10 points)

```python
#codes
grouped_df = happiness_df.groupby("happiness_degree")
mean_df = grouped_df.mean()
mean_df
```

Question 9) Plot multiple bar charts in a single chart that show the mean values of 'Social support','Freedom','Generosity' according to happiness degree (15 points)

```python
#codes
N = 5
ind = np.arange(N) 
width = 0.15
  

xvals = [happiness_df.groupby('happiness_degree')['Social support'].mean()['very happy'],
happiness_df.groupby('happiness_degree')['Social support'].mean()['happy'],
happiness_df.groupby('happiness_degree')['Social support'].mean()['moderately happy'],
happiness_df.groupby('happiness_degree')['Social support'].mean()['unhappy'],
happiness_df.groupby('happiness_degree')['Social support'].mean()['very unhappy']]
bar1 = plt.bar(ind, xvals, width, color = 'r')

yvals =[happiness_df.groupby('happiness_degree')['Freedom'].mean()['very happy'],
happiness_df.groupby('happiness_degree')['Freedom'].mean()['happy'],
happiness_df.groupby('happiness_degree')['Freedom'].mean()['moderately happy'],
happiness_df.groupby('happiness_degree')['Freedom'].mean()['unhappy'],
happiness_df.groupby('happiness_degree')['Freedom'].mean()['very unhappy']]
bar2 = plt.bar(ind+width, yvals, width, color='g')
  
zvals = [happiness_df.groupby('happiness_degree')['Generosity'].mean()['very happy'],
happiness_df.groupby('happiness_degree')['Generosity'].mean()['happy'],
happiness_df.groupby('happiness_degree')['Generosity'].mean()['moderately happy'],
happiness_df.groupby('happiness_degree')['Generosity'].mean()['unhappy'],
happiness_df.groupby('happiness_degree')['Generosity'].mean()['very unhappy']]
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')
  



plt.title(" The mean values of Social support,Freedom,Generosity according to Happiness Degree")
  
plt.xticks(ind+width,['very happy','happy','moderately happy','unhappy','very unhappy'],fontsize=11)
plt.legend( (bar1, bar2, bar3), ('Social support','Freedom','Generosity'),fontsize=13)
plt.show()
```
